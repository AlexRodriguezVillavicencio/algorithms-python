```python
entrada = input('nombre del archivo? ')
archivo = open(entrada, 'r')
```

```python
c = archivo.read()
```

```python
archivo.close()
```

```python
c = c.lower()
```

```python
dic2 = {}
for key in c:
    if key not in dic2:
        dic2[key] = 1
    else:
        dic2[key] +=1
```

```python
dic2
```

```python
for key, value in dic2.items():
    print(key, '->', value)
```

```python
sortedDict = sorted(dic2.items(), key=lambda x: x[1], reverse = True)
sortedDict
```

```python
for key in sortedDict:
    print(key[0], '->', key[1])
```
